```python
#main 10 attributes,others are derived from it. Considering mean here
cols = ['diagnosis',
        'radius_mean', 
        'texture_mean', 
        'perimeter_mean', 
        'area_mean', 
        'smoothness_mean', 
        'compactness_mean', 
        'concavity_mean',
        'concave points_mean', 
        'symmetry_mean', 
        'fractal_dimension_mean']
sns.pairplot(data=df[cols], hue='diagnosis', palette='RdBu')
```

```python
#corr.style.background_gradient(cmap='coolwarm').set_precision(2)
corr = df.corr().round(2)
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
f, ax = plt.subplots(figsize=(20,20))
sns.heatmap(corr, mask=mask, cmap='coolwarm', vmin=-1, vmax=1, center=0,square=True,
             linewidths=.5,cbar_kws={"shrink":.5},annot=True)
plt.tight_layout()
```

```python
cols = ['radius_worst', 
        'texture_worst', 
        'perimeter_worst', 
        'area_worst', 
        'smoothness_worst', 
        'compactness_worst', 
        'concavity_worst',
        'concave points_worst', 
        'symmetry_worst', 
        'fractal_dimension_worst',
        'perimeter_mean',
        'perimeter_se', 
        'area_mean', 
        'area_se',
        'concavity_mean',
        'concavity_se', 
        'concave points_mean', 
        'concave points_se']
df = df.drop(cols, axis=1)
df.columns
```

```python
corr = df.corr().round(2)
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
f, ax = plt.subplots(figsize=(20,20))
sns.heatmap(corr, mask=mask, cmap='coolwarm', vmin=-1, vmax=1, center=0,square=True,
             linewidths=.5,cbar_kws={"shrink":.5},annot=True)
plt.tight_layout()
```

```python
df.head()
```

```python
df['radius_mean']=(df['radius_mean']-df['radius_mean'].mean())/(df['radius_mean'].std())
```

```python
df['texture_mean']=(df['texture_mean']-df['texture_mean'].mean())/(df['texture_mean'].std())
```

```python
df.head(10)
```

```python
y=df['diagnosis']
x=df.drop('diagnosis',axis='columns')
```

```python
x.head()
```

```python
y=np.array(y.values.reshape((len(df),1)))
```

```python
y.shape
```

```python
x=np.array(x)
x.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3,random_state=40)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_predicted=model.predict(X_test)
```

```python
y_predicted 
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
plt.figure(figsize = (2,2))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
true_negative = cm[0][0]
false_positive = cm[0][1]
false_negative = cm[1][0]
true_positive = cm[1][1]
print('True Negative:', true_negative)
print('False Positive:', false_positive)
print('False Negative:', false_negative)
print('True Positive:', true_positive)
print('Correct Predictions', round((true_negative + true_positive) / len( y_predicted) * 100, 1), '%')
```
